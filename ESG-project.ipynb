```python
def boxplot(file, column): 
    plt.figure(figsize=(12,6))
    sns.boxplot(x="Industry", y=column, data=file, color='skyblue')
    plt.title(f'{column} Scores by Industry')
    plt.xticks(rotation=45)
    return plt.show()

boxplot(data, 'ESG_Overall')
```

```python
# Удаление выбросов 
new_df = data.copy()
def remove_outliers(file, industrial_column, esg_column):
    copy = file.loc[data['Industry'] == industrial_column, :]
    copy = copy.sort_values(by = esg_column)
    
    Q1 = copy[esg_column].quantile(0.25)
    Q3 = copy[esg_column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outlier_indices = copy[(copy[esg_column] < lower_bound) | (copy[esg_column] > upper_bound)].index
    new = file.drop(outlier_indices)
    return new 
```

```python
new_df = remove_outliers(new_df, 'Consumer Goods', 'ESG_Overall')
new_df = remove_outliers(new_df, 'Retail', 'ESG_Overall')
new_df = remove_outliers(new_df, 'Utilities', 'ESG_Overall')
boxplot(new_df, 'ESG_Overall')
```

```python
for industry in industries: 
    new_df = remove_outliers(new_df, industry,'ProfitMargin')
for industry in industries: 
    new_df = remove_outliers(new_df, industry,'GrowthRate')
for industry in industries: 
    new_df = remove_outliers(new_df, industry,'Revenue')
```

```python
def is_normal(group,alpha): 
    ks_test = stats.kstest(group, 'norm', args=(np.mean(group), np.std(group, ddof=1)))
    return ks_test.pvalue > alpha

def kdeplot(groupA, groupB, industry): 
    sns.set_style("whitegrid")
    plt.figure(figsize=(10, 6))

    sns.kdeplot(groupA, label='Group A', color='skyblue', fill=True, alpha=0.3, linewidth=2)
    sns.kdeplot(groupB, label='Group B', color='orange', fill=True, alpha=0.3, linewidth=2)
    
    plt.axvline(np.mean(groupA), color='blue', linestyle='--', linewidth=1)
    plt.axvline(np.mean(groupB), color='orange', linestyle='--', linewidth=1)
    
    plt.xlabel(industry, fontsize=12)
    plt.ylabel('density', fontsize=12)
    plt.legend(title='Group', loc='upper right')
    plt.tight_layout()
    return plt.show()


def test(industry,depend, independ): 
    median = new_df[new_df['Industry'] == industry][independ].median()
    groupA = new_df[(new_df['Industry'] == industry) & (new_df[independ] > median)][depend]
    groupB = new_df[(new_df['Industry'] == industry) & (new_df[independ] <= median)][depend]


    if is_normal(groupA, 0.05) and is_normal(groupB, 0.05): 
        # Т-критерий Вилкоксона для независимых выборок
        t_stat, p_value = ttest_ind(groupA, groupB, equal_var=False)  
        res = 'не влияет' if p_value > 0.05 else 'влияет'
    else: 
        # Критерий Манна-Уитни
        m_stat, p_value = stats.mannwhitneyu(groupA, groupB)
        res = 'не влияет' if p_value > 0.05 else 'влияет'
        
    kdeplot(groupA, groupB, industry)
    
    return res
```

### A/B тест №1: “Влияние отрасли компании на рост ESG”

Н0: Отрасль компании не оказывает статистически значимого влияния на рост её ESG-показателей.

Н1: Рост ESG-показателей зависит от отрасли компании.

В исследовании участвуют четыре принципиально разные отрасли, специально отобранные для максимального контраста в анализе:

Finance, Technology, Energy, Consumer Goods

```python
selected_industries = ['Finance', 'Energy', 'Technology', 'Consumer Goods']

test1 = new_df[["Industry", "ESG_Overall"]].dropna()
df_balanced = test1.groupby("Industry").sample(n=95, random_state=42, replace=True)
new_test1 = pd.DataFrame()

for industry in selected_industries:
    new_test1[industry] = df_balanced[df_balanced['Industry'] == industry]['ESG_Overall'].values
new_test1
```

```python
for i in new_test1.columns:
    print(f'Нормальность {i}: {is_normal(new_test1[i],0.05)}')
```

```python
# ANOVA 
f_stat, p_value = f_oneway( new_test1['Energy'], new_test1['Finance'], 
                        new_test1['Technology'], new_test1['Consumer Goods'])
print(f"F-статистика: {f_stat}, p-value: {p_value}")
```

```python
'Отрасли не влияют на ESG' if p_value > 0.05 else 'Отрасли влияют на ESG'
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

for industry in selected_industries:
    sns.kdeplot(new_test1[industry], label=industry, fill=True, alpha=0.3, linewidth=1.5)

plt.xlabel('values', fontsize=12)
plt.ylabel('density', fontsize=12)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### A/B тест №2: “Влияние ESG на прибыльность компании (ProfitMargin)"

Н0: ESG-показатели компании не оказывают статистически значимого влияния на её прибыльность.

Н1: Компании с высокими ESG демонстрируют более высокую прибыльность по сравнению с компаниями с низкими ESG.

```python
test2 = pd.DataFrame(columns=['Industry','ESG на ProfitMargin'])
for industry in industries:
    test2_industry = test(industry,'ProfitMargin', 'ESG_Overall')
    row2 = {'Industry' : industry, 'ESG на ProfitMargin':test2_industry}
    test2 = pd.concat([test2, pd.DataFrame([row2])], axis=0, ignore_index= True)

test2
```

### A/B тест №3: “Влияние ESG на рост компании (GrowthRate)"

Н0: ESG-показатели компании не оказывают статистически значимого влияния на её рост.

Н1: Компании с высокими ESG демонстрируют более высокий рост по сравнению с компаниями с низкими ESG.

```python
test3 = pd.DataFrame(columns=['Industry','ESG на GrowthRate'])
for industry in industries:
    test3_industry = test(industry,'GrowthRate','ESG_Overall')
    row3 = {'Industry' : industry, 'ESG на GrowthRate':test3_industry}
    test3 = pd.concat([test3, pd.DataFrame([row3])], axis=0, ignore_index= True)

test3
```

#### Вывод: 

Проведенный анализ демонстрирует, что ESG не оказывают значимого влияния на ключевые финансовые показатели — прибыль и рост компании — в большинстве отраслей. Корреляция между высокими ESG-рейтингами и улучшением финансовых результатов статистически незначима, что указывает на вторичную роль устойчивого развития в краткосрочной доходности бизнеса.

Однако отрасль Utilities становится исключением из этого правила. Здесь наблюдается обратная зависимость: снижение ESG-показателей может сопровождаться ростом прибыли.
